# Training Data Generation

```python
expressionsDict = {
    "e0" : [["not",["Unterschrank(z)","and",["not","Moebel(z)"]]], 2],
    "e1" : [["not",["hatLeistungserbringer(x,y)","and",["versandterBeleg(y,x)","and",["not","Ausgangsrechnung(x)"]]]], 3]
}
```

```python
from tnreason.model import sampling
```

```python
sampler = sampling.GibbsSampler(expressionsDict)
```

```python
sampleDf = sampler.create_sampleDf(100)
```

```python
sampleDf.head()
```

# Learn Formulas

```python
from tnreason.learning import mle_integrated_learning as mlel
```

```python
learner = mlel.FormulaLearner(sampleDf=sampleDf)
```

```python
#skeletonExpression = ["not",["P1","and",["not","P2"]]]
skeletonExpression = ["P1","and","P2"]
candidatesDict = {
    "P1" : ["Unterschrank(z)","Moebel(z)", "Ausgangsrechnung(x)"],
    "P2" : ["Unterschrank(z)","Moebel(z)", "Ausgangsrechnung(x)"]
}
```

```python
learner.learn(skeletonExpression, candidatesDict, gdSweepNum = 100)
```
