# Inference of Hybrid Knowledge Bases

A hybrid Knowledge Base consists of
- logical facts (factsDict): Hard constraints on the possible worlds.
- probabilistic formulas (weightedFormulasDict): Modelling the probability of possible worlds.

```python
from tnreason import knowledge
```

```python
weightedFormulasDict = {
    "formula0" : [["a1","imp","a2"], 2],
    "formula1" : ["a1", 2],
    "formula2" : [["a2","eq",["not","a3"]], -1]
}

factsDict = {
    "fact1" : ["a2","eq",["not","a4"]]
}
```

```python
hybridKB = knowledge.HybridKnowledgeBase(
    weightedFormulasDict,
    factsDict
)
```

```python
hybridKB.visualize()
```

## Inference by annealing algorithms


```python
hybridKB.annealed_map_query(variableList=["a3","a4"], evidenceDict= {"a2": 0})
```

```python
hybridKB.exact_map_query(variableList=["a3","a4"], evidenceDict= {"a2": 0})
```

```python
hybridKB.query(variableList=["a3"], evidenceDict = {"a2":0}).values
```

```python
hybridKB.ask("a2")
```

```python
hybridKB.ask(["not","a2"])
```

```python
hybridKB.ask_constraint(["a2","eq",["not","a4"]])
```

```python
hybridKB.evaluate_evidence({"a1":1, "a2":0})
```
