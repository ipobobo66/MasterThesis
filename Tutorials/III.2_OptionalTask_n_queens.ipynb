The n-queens problem consists in the assignments of n-queens on a nxn chess board, such that no pair of queens is in the same column or in the same row.

Task: Use the library tnreason to model the n-queens problem (Hint: Orient on the Sudoku tutorial and ).

# Task: Model the categorical constraints
