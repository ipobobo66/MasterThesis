# Generation of Data

```python
from tnreason import knowledge

hybridKB = knowledge.HybridKnowledgeBase(
    facts = {"f1": ["imp","Rained","Street"]},  #the rain will hit the street with 1 probability
    weightedFormulas= {"w1" : ["imp","Sprinkler","Street",0.5]} #the rain will hit the street with .5 probability
)
```

```python
inferer = knowledge.InferenceProvider(hybridKB)
```

```python
samples = inferer.draw_samples(200)
```

```python
samples
```

# Learning the formula

```python
architecture = {
    "neur1" : [["xor","and","imp"],
               ["Rained","Sprinkler"],
               ["Street"]
               ],
}
```

```python
booster = knowledge.FormulaBooster(knowledgeBase = knowledge.HybridKnowledgeBase(),
                                   specDict = {
                                     "method" : "als",
                                     "sweeps" : 10,
                                     "headNeurons" : ["neur1"],
                                     "architecture": architecture
                                   })
```

```python
booster.find_candidate(sampleDf=samples)
```
