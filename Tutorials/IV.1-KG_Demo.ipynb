# Create a Knowledge Graph from the Protege THWS Demo Ontology

https://webprotege.stanford.edu/\#projects/db1f026b-b34c-4955-8cd0-d5dc167098dc/edit/

We include the ontology as a turtle string in the following and perform queries.

```python
ontology_string = """@prefix : <urn:webprotege:ontology:13ff9d95-f8ca-43c1-8877-5fcebb2180e2#> .
@prefix loc: <www.locationdemo.de/location/ontology#> .
@prefix owl: <http://www.w3.org/2002/07/owl#> .
@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .
@prefix xml: <http://www.w3.org/XML/1998/namespace> .
@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .
@prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> .
@base <urn:webprotege:ontology:13ff9d95-f8ca-43c1-8877-5fcebb2180e2> .

<urn:webprotege:ontology:13ff9d95-f8ca-43c1-8877-5fcebb2180e2> rdf:type owl:Ontology .

#################################################################
#    Object Properties
#################################################################

###  www.locationdemo.de/location/ontology#hasCampusAt
loc:hasCampusAt rdf:type owl:ObjectProperty ;
                rdfs:subPropertyOf owl:topObjectProperty ;
                rdfs:domain loc:University ;
                rdfs:range loc:Location ;
                rdfs:label "loc:hasCampusAt" .


###  www.locationdemo.de/location/ontology#hasCampusIn
loc:hasCampusIn rdf:type owl:ObjectProperty ;
                rdfs:subPropertyOf owl:topObjectProperty ;
                rdfs:domain loc:University ;
                rdfs:range loc:Location ;
                rdfs:label "loc:hasCampusIn" .


###  www.locationdemo.de/location/ontology#locatedIn
loc:locatedIn rdf:type owl:ObjectProperty ;
              rdfs:subPropertyOf owl:topObjectProperty ;
              rdfs:domain loc:City ;
              rdfs:range loc:Country ;
              rdfs:label "loc:locatedIn" .


###  www.locationdemo.de/location/ontology#studiesInCity
loc:studiesInCity rdf:type owl:ObjectProperty ;
                  rdfs:subPropertyOf owl:topObjectProperty ;
                  rdfs:domain loc:Student ;
                  rdfs:range loc:City ;
                  rdfs:label "loc:studiesInCity" .


###  www.locationdemo.de/location/ontology#studiesSubject
loc:studiesSubject rdf:type owl:ObjectProperty ;
                   rdfs:subPropertyOf owl:topObjectProperty ;
                   rdfs:domain loc:Student ;
                   rdfs:range loc:StudySubject ;
                   rdfs:label "loc:studiesSubject" .


#################################################################
#    Classes
#################################################################

###  www.locationdemo.de/location/ontology#AppliedScienceSubject
loc:AppliedScienceSubject rdf:type owl:Class ;
                          rdfs:subClassOf loc:StudySubject ;
                          rdfs:label "loc:AppliedScienceSubject" .


###  www.locationdemo.de/location/ontology#City
loc:City rdf:type owl:Class ;
         rdfs:subClassOf loc:Location ;
         rdfs:label "loc:City" .


###  www.locationdemo.de/location/ontology#ComputerScienceSubject
loc:ComputerScienceSubject rdf:type owl:Class ;
                           rdfs:subClassOf loc:StudySubject ;
                           rdfs:label "loc:ComputerScienceSubject" .


###  www.locationdemo.de/location/ontology#Country
loc:Country rdf:type owl:Class ;
            rdfs:subClassOf loc:Location ;
            rdfs:label "loc:Country" .


###  www.locationdemo.de/location/ontology#Location
loc:Location rdf:type owl:Class ;
             rdfs:label "loc:Location" .


###  www.locationdemo.de/location/ontology#Student
loc:Student rdf:type owl:Class ;
            rdfs:label "loc:Student" .


###  www.locationdemo.de/location/ontology#StudySubject
loc:StudySubject rdf:type owl:Class ;
                 rdfs:label "loc:StudySubject" .


###  www.locationdemo.de/location/ontology#University
loc:University rdf:type owl:Class ;
               rdfs:label "loc:University" .


###  Generated by the OWL API (version 4.5.13) https://github.com/owlcs/owlapi

"""
```

```python
kg = kglab.KnowledgeGraph()
kg.load_rdf_text(ontology_string)
```

# Adding some examples and query the Knowledge Graph

```python
examples_string = """
@prefix loc: <www.locationdemo.de/location/ontology#> .
@prefix owl: <http://www.w3.org/2002/07/owl#> .
@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .
@prefix xml: <http://www.w3.org/XML/1998/namespace> .
@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .
@prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> .

loc:Alice rdf:type loc:Student ;
    loc:studiesSubject loc:TechnoMathematik .

loc:Bob rdf:type loc:Student ;
    loc:studiesSubject loc:ArtificialIntelligence .

loc:TechnoMathematik rdf:type loc:AppliedScienceSubject .

loc:ArtificialIntelligence rdf:type loc:ComputerScienceSubject .

loc:THWS rdf:type loc:University ;
    loc:hasCampusIn loc:Schweinfurt .

loc:Schweinfurt rdf:type loc:City ;
    loc:locatedIn loc:Germany .

loc:Germany rdf:type loc:Country .

"""

kg.load_rdf_text(examples_string)
```

```python
sparql_query = """
SELECT ?student ?subject ?subjectType
WHERE {
    ?student rdf:type loc:Student .
    ?student loc:studiesSubject ?subject .
    ?subject rdf:type ?subjectType .
}
"""

resultDf = kg.query_as_df(sparql_query)
resultDf
```

# Inference Example: Built-in RDFS rule

We now want to infer all axioms, which are entailed by the above axioms.

In this example, we demonstrate that by inference the objects "loc:Schweinfurt" and "loc:Germany" are infered to be instances of the class "loc:Location". This is because they we defined them as instances of "loc:City" and "loc:Country", which are sub-classes (defined by "rdfs:subClassOf") of "loc:Locations".



We initialize a Knowledge Graph "inference_kg", and show, that there is no individual of type "loc:Location".

```python
inference_kg = kglab.KnowledgeGraph()
inference_kg.load_rdf_text(ontology_string)
inference_kg.load_rdf_text(examples_string)
```

```python
location_query = """
SELECT ?object ?location
WHERE {
  ?object rdf:type loc:Location .
}
"""

resultDf = inference_kg.query_as_df(location_query)
resultDf
```

After performing inference on the Knowledge Graph, there are now individuals of type "loc:Location".

```python
inference_kg.infer_owlrl_closure()

resultDf = inference_kg.query_as_df(location_query)
resultDf
```

# Inference Example: Hand-made Role Inclusion Axiom

We now want to add the knowledge to the graph, that any university having a campus in Schweinfurt has also a campus in Germany.


First, we demonstrate, that inference based on the Knowledge Graph initialized with the above strings does not capture this rule.

```python
inference_kg = kglab.KnowledgeGraph()
inference_kg.load_rdf_text(ontology_string)
inference_kg.load_rdf_text(examples_string)

inference_kg.infer_owlrl_closure()
```

```python
campus_query = """
SELECT ?university ?location
WHERE {
    ?university rdf:type loc:University .
    ?university loc:hasCampusIn ?location .
}
"""

resultDf = inference_kg.query_as_df(campus_query)
resultDf
```

We add the rule, encoded in a "owl:propertyChainAxiom" and infer again.

```python
rules = """
@prefix loc: <www.locationdemo.de/location/ontology#> .
@prefix owl: <http://www.w3.org/2002/07/owl#> .
@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .
@prefix xml: <http://www.w3.org/XML/1998/namespace> .
@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .
@prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> .

loc:hasCampusIn owl:propertyChainAxiom (loc:hasCampusIn loc:locatedIn) .

"""
inference_kg.load_rdf_text(rules)
```

```python
inference_kg.infer_owlrl_closure()

resultDf = inference_kg.query_as_df(campus_query)
resultDf
```

# Inference Based on Rules in SPARQL

```python
rdf_kg = rdflib.Graph()
rdf_kg.parse(data = ontology_string)
rdf_kg.parse(data = examples_string)
```

```python
schweinfurt_query = """
SELECT ?student
WHERE {
  ?student loc:studiesInCity loc:Schweinfurt .
}
"""
result = rdf_kg.query(schweinfurt_query)
for row in result:
  print(f"{row.student}")
```

```python
insert_query = """
INSERT {
  ?student loc:studiesInCity loc:Schweinfurt .
}
WHERE {
  ?student loc:studiesSubject ?subject .
  ?subject rdf:type loc:AppliedScienceSubject .
}
"""
rdf_kg.update(insert_query)
```

```python
schweinfurt_query = """
SELECT ?student
WHERE {
  ?student loc:studiesInCity loc:Schweinfurt .
}
"""
result = rdf_kg.query(schweinfurt_query)
for row in result:
  print(f"{row.student}")
```
